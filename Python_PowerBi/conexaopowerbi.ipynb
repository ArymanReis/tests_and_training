```python
import pandas as pd;
import matplotlib as mp;
import seaborn as sn;
```

```python
dataframe = pd.read_csv('Invictus_faturamento_vs_margem_2024.csv')


dataframe.head(n=10)
```

```python
print(dataframe['DATA'].unique())
```

```python
meses = {
    1: 'Janeiro',
    2: 'Fevereiro',
    3: 'Março',
    4: 'Abril',
    5: 'Maio',
    6: 'Junho',
    7: 'Julho',
    8: 'Agosto',
    9: 'Setembro',
    10: 'Outubro',
    11: 'Novembro',
    12: 'Dezembro'
}

# Substituir números pelos nomes dos meses
dataframe['DATA'] = dataframe['DATA'].map(meses)
```

```python
dataframe.head()
```

```python
# Função para remover caracteres especiais
def remover_caracteres(valor):
    if isinstance(valor, str):  # Verifica se o valor é uma string
        valor = valor.strip()  # Remove espaços extras
        
        # Remove símbolo de moeda e separadores de milhar
        valor = valor.replace('$', '').replace(',', '').replace(' ', '')

        return valor  # Retorna o valor como string ainda (não converte para float)

    return valor  # Retorna o valor original se não for string

# Lista das colunas monetárias
colunas_monetarias = ['FATURAMENTO ORÇADO (R$)', 'FATURAMENTO REAL(R$)', 'FECHAMENTO ORÇADO (R$)', 'FECHAMENTO REAL (R$)']

# Aplicar a função para remover caracteres especiais
for coluna in colunas_monetarias:
    dataframe[coluna] = dataframe[coluna].apply(remover_caracteres)

# Visualizar as primeiras linhas para verificar o resultado
print(dataframe.head())
```

```python
dataframe.info()
```

```python
# Função para converter valores para float
def converter_para_float(valor):
    if isinstance(valor, str):  # Verifica se ainda é uma string
        valor = valor.strip()  # Remove espaços extras
        
        # Verifica se o número está no formato contábil negativo (exemplo: "(16682.31)")
        if valor.startswith('(') and valor.endswith(')'):
            valor = '-' + valor[1:-1]  # Remove os parênteses e adiciona o sinal de negativo

        try:
            return float(valor)  # Converte para float
        except ValueError:
            print(f"Erro ao converter valor: {valor}")  # Mensagem de erro para depuração
            return None  # Retorna None caso a conversão falhe

    return valor  # Retorna o valor original se não for string

# Aplicar a conversão para float nas colunas monetárias
for coluna in colunas_monetarias:
    dataframe[coluna] = dataframe[coluna].apply(converter_para_float)

# Verificar se a conversão deu certo
print(dataframe.info())  # Agora as colunas devem aparecer como float64
print(dataframe.head())  # Visualizar os primeiros valores convertidos
```
